```python
df = pd.read_csv("../data/data_index_per_fly.csv")
df = df.sort_values(by="Time of death (hours into video)", ascending=False).reset_index(
    drop=True
)
df["Fly id"] = df["Fly id"].astype(str)
df = df[df["Time of death (hours into video)"] > 0]
```

```python
import os
print(os.getcwd())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# use textwrap from python standard lib to help manage how the description
# text shows up
import textwrap
```

```python
pal = sns.cubehelix_palette(df.shape[0], rot=-0.25, light=0.7)
print(pal.as_hex())
pal
```

```python
# plt.style.use("seaborn-whitegrid")  # set style because it looks nice
# plt.style.use("seaborn-white")  # set style because it looks nice
plt.style.use("default")  # set style because it looks nice

fig, ax = plt.subplots(1, 1, figsize=(3.75, 3.75), dpi=600)

# color palette to choose from
darkblue = "#1f253f"
lightblue = "#58849f"
redish = "#d73027"

DOT_SIZE = 50

# create the various dots
# avg dot
ax.scatter(
    x=df["Time of death (hours into video)"],
    y=df["Fly id"].astype(str),
    s=DOT_SIZE,
    alpha=1,
    label=None,
    color=pal,
    edgecolors="white",
)


x_min, x_max = ax.get_xlim()
y_min, y_max = ax.get_ylim()

# ax.plot([df["Time of death (hours into video)"].mean(),df["Time of death (hours into video)"].mean()], [y_min, y_max], linestyle="-", color="white", linewidth=3,alpha=0.4)
ax.plot(
    [
        df["Time of death (hours into video)"].mean(),
        df["Time of death (hours into video)"].mean(),
    ],
    [y_min, y_max],
    linestyle="--",
    color=redish,
    alpha=0.4,
)

ax.text(
    x=df["Time of death (hours into video)"].mean() - 1,
    y=(y_min) + 0.5,
    s="Mean lifespan",
    horizontalalignment="right",
    verticalalignment="bottom",
    fontsize=8,
    color=redish,
    rotation="vertical",
    # backgroundcolor="white",
    alpha=0.4,
)
ax.tick_params(color="black", labelcolor="black")
for spine in ax.spines.values():
    spine.set_edgecolor("black")

# remove the y ticks
ax.set_yticks([])

ax.grid(False)
ax.spines["top"].set_visible(False)
# ax.spines["bottom"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.spines["left"].set_visible(False)

ax.set_xticks([0, 24, 48, 72, 96, 120, 144, 168, 192])
ax.set_xlabel("Hours tracked (time of death) ", fontsize=8, color="black")

l = ax.legend(
    frameon=False,
    bbox_to_anchor=(0.6, 1.05),
    ncol=2,
    fontsize="x-large",
    labelcolor="black",
)


for text in l.get_texts():
    text.set_color("dimgrey")
ax.hlines(
    y=df["Fly id"],
    xmin=0,
    xmax=df["Time of death (hours into video)"],
    color="black",
    alpha=0.4,
    lw=1,
    zorder=0,
)

plt.margins(x=0, y=0)
plt.xlim(0, 200)
plt.savefig("figures/lifespans.svg", dpi=600, bbox_inches="tight")
plt.show()
```
